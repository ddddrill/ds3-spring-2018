График 1
Какое колличество жанров выпускалось в разные года

```python
%pylab inline                      
from pylab import rcParams         # дефолтный размер графиков увеличенен        
rcParams['figure.figsize']=6,3

import matplotlib as plt
import seaborn as sns

data=data.dropna()                 # убираем строки в которых ничего нет
data.head()

```

```python
'''Создадим график кол-ва жанров выпускаемых игр на каждый год'''

genre=data['Genre'].to_list()    # сделали список из всех жанров 
genre=set(genre)                # выделили уникальные жанры из всего набора

'''Вводим нужный нам год '''
year=1999                      #введите сюда нужный год, для которого вам необходима статистика

g_y=data.loc[data['Year_of_Release']==year]  # выделяем только интересующий нас год
```

```python
'''Создадим DataFrame на основе которого будем пользоваться графиком'''
value_genre=g_y['Genre'].value_counts()
year_genre=pd.DataFrame(data=value_genre)   # DataFrame в котором содержиться общее кол-во каждого жанра за год
```

```python
'''Начнем визуализацию с помощью matplotlib'''
pl=year_genre.plot(kind='bar',rot=45)
```
